```python
#Importing the required libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading all the datasets

user_details = pd.read_excel("Data Analyst Intern Assignment - Excel.xlsx", sheet_name = 'UserDetails.csv')
```

```python
user_details
```

```python
cooking_sessions = pd.read_excel("Data Analyst Intern Assignment - Excel.xlsx", sheet_name = 'CookingSessions.csv')
cooking_sessions
```

```python
order_details = pd.read_excel("Data Analyst Intern Assignment - Excel.xlsx", sheet_name = 'OrderDetails.csv')
order_details
```

# Data Preprocessing

```python
# Checking for missing values in all columns

print(user_details.isnull().sum())
```

```python
print(cooking_sessions.isnull().sum())
```

```python
order_details.isnull().sum()
```

```python
#Handling missing values in Rating column
order_details['Rating'] = order_details['Rating'].fillna(0)
```

```python
order_details.isnull().sum()
```

```python
# Checking for duplicate rows

print(user_details.duplicated().sum())
print(cooking_sessions.duplicated().sum())
print(order_details.duplicated().sum())
```

```python
# Merging UserDetails with CookingSessions
merged_data = pd.merge(user_details, cooking_sessions, on='User ID', how='inner')
```

```python
merged_data
```

```python
final_data = pd.merge(merged_data, order_details, on='User ID')
final_data
```

```python
final_data.describe()
```

# Visualizations using EDA

```python
#plotting total order distribution according to age
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='Total Orders', data=final_data)
plt.title('Total Orders by Age')
plt.xlabel('Age')
plt.ylabel('Total Orders')
plt.show()
```

```python
# most popular dishes
popular_dishes = final_data['Dish Name_y'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=popular_dishes.index, y=popular_dishes.values)
plt.title('Dishes according to popularity')
plt.xlabel('Dish Name')
plt.ylabel('Number of Orders')

plt.show()
```

```python
# Scatter plot of session duration vs total orders
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration (mins)', y='Total Orders', data=final_data)
plt.title('Session Duration vs Total Orders')
plt.xlabel('Session Duration (minutes)')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Scatter plot of user name vs total orders
plt.figure(figsize=(10, 6))
sns.scatterplot(x='User Name', y='Total Orders', data=final_data)
plt.title('User name vs Total Orders')
plt.xlabel('User Name')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Barplot of Total Orders according to meal type
plt.figure(figsize=(10, 6))
sns.barplot(x='Meal Type_y', y='Total Orders', data=final_data)
plt.title('Total Orders according to meal')
plt.xlabel('Meal type')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Meal Type Distribution by Age
plt.figure(figsize=(10, 6))
sns.countplot(x='Age', hue='Meal Type_y', data=final_data)
plt.title('Meal Type Preferences by Age')
plt.xlabel('Age')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Dishes ordered according to time of the day
plt.figure(figsize = (10,6))
sns.countplot(x="Time of Day", hue='Dish Name_y', data=final_data)
plt.title('Dishes ordered according to time')
plt.xlabel('Time of Day')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Number of orders placed according to time of the day
plt.figure(figsize=(10, 6))
sns.barplot(x="Time of Day", y='Total Orders', data=final_data)
plt.title('Total Orders according to the time')
plt.xlabel('Time of Day')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Dish Preferences by Age

plt.figure(figsize=(10, 6))
sns.countplot(x='Age', hue='Dish Name_y', data=final_data)
plt.title('Dish Preferences by Age')
plt.xlabel('Age')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Bar plot of total orders by location
plt.figure(figsize=(12, 6))
sns.barplot(x='Location', y='Total Orders', data=final_data)
plt.title('Total Orders by Location')
plt.xlabel('Location')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot of order amount by time of day
plt.figure(figsize=(12, 6))
sns.boxplot(x='Time of Day', y='Amount (USD)', data=final_data)
plt.title('Order Amount by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Order Amount (USD)')
plt.show()
```
